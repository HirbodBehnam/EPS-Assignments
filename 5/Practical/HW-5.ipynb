# dealing with NA's

In R, NA stands for Not Available. Each cell of your data that displays NA is a missing value. there are several ways for dealing with them, like omit those rows or fill them with relative data's like mean or zero or median.
now in this part we want you to fill na values of column 'car' with median and other columns with mean and call summary function again.

```r
# https://stackoverflow.com/a/53346070/4213397
buyers <- buyers %>% mutate_at(vars(Cars), ~ifelse(is.na(.x), median(.x, na.rm = TRUE), .x))
buyers <- buyers %>% mutate_all(~ifelse(is.na(.x), mean(.x, na.rm = TRUE), .x))
print(summary(buyers))
```

## Bar Plot
A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. A bar plot shows comparisons among discrete categories. One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.

```r
# I'll remove all empty ones
# I can also use  %>% mutate_at(vars(Region), ~ifelse(.x == "", "NA", .x))
# to create a bar for them as well
buyers_temp <- buyers[buyers$Region != "",]
print(ggplot(buyers_temp, aes(Region)) + geom_bar())
```

## Scatter Plot
A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```r
# Just like salaries we can find out income to age and gender
print(ggplot(buyers, aes(Age, Income)) +
    geom_point(aes(color = Gender)))
```

## Histogram
A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range. A histogram displays the shape and spread of continuous sample data.

```r
# Just a chart for Age
ggplot(buyers, aes(x = Age)) + geom_histogram(colour = "black", fill = "white")
```

## Density Plot
A density plot is a representation of the distribution of a numeric variable. It uses a kernel density estimate to show the probability density function of the variable
. It is a smoothed version of the histogram and is used in the same concept.

```r
# Just like above!
# http://www.sthda.com/english/wiki/ggplot2-histogram-plot-quick-start-guide-r-software-and-data-visualization
ggplot(buyers, aes(x = Age)) + 
    geom_histogram(aes(y = ..density..), colour = "black", fill = "white") +
    geom_density(alpha = .2, fill = "red")
```

## Trend Plot
Trend plots are used to illustrate how a random variable may change over time (e.g. sales, exchange rates, stick prices). In order to create a Trend plot one must first have defined a set of cells as an input or output array.

```r
ggplot(salaries, aes(yrs.since.phd, salary)) +
geom_point(aes(colour = rank)) +
geom_smooth(aes(colour = rank))
```

## Introduction
So I found this datasheet on internet which is about prices and specs of houses in King County, USA. It's free and can be downloaded from [here](https://www.kaggle.com/datasets/harlfoxem/housesalesprediction). It's also included in the datasheet folder.

I wanted to see if the relation between house area (living area) and price is linear.

So at first I wrote a code to load the dataframe and put in into scatter plot.

```r
library(ggplot2)
library(dplyr)
# CSV from https://www.kaggle.com/datasets/karthikbhandary2/property-rentals
rentals <- read.csv("datasets/kc_house_data.csv")
summary(rentals)
print(ggplot(rentals, aes(sqft_living, price)) + geom_point())
```

It looks horrible. There are some realy big houses and because of the presense of a lot of mid-level houses, points create a big black mole in buttom corner.

So we have to reduce the dataset size. The dataset has more than 20K entries. Let's remove random elements from it to have only 2000 enties at last.

```r
reduced_rentals <- sample_n(rentals, 2000)
print(ggplot(reduced_rentals, aes(sqft_living, price)) + geom_point())
```

Way better.

But I don't like the big range of price. Let's use log to reduce it.

```r
rentals_log_price <- reduced_rentals %>% mutate_at(vars(price), ~log(.x))
print(ggplot(rentals_log_price, aes(sqft_living, price)) + geom_point())
```

Now data looks very cleaner. For the last step, lets draw the trend line.

```r
print(ggplot(rentals_log_price, aes(sqft_living, price)) + geom_point() + geom_smooth())
```

It feels like that this line is linear.

***BUT*** this line is linear is a logarithmic scale plot. This means that in real plot, this line would have been exponential!

## Spicing things up
Now let's see how much renovation effects the price. The datasheet contains the renovation year. If the cell data is zero, it means that there was not renovation but other values indicate the renovation date.

Again, lets choose 2000 random houses and find out if they have been rebuilt or not. I'll also limit the range of living area.

```r
rentals_for_rebuilt <- sample_n(rentals[rentals$sqft_living < 6000,], 2000)
rentals_for_rebuilt$rebuilt <- rentals_for_rebuilt$yr_renovated != 0
rentals_for_rebuilt <- rentals_for_rebuilt %>% mutate_at(vars(price), ~log(.x))
ggplot(rentals_for_rebuilt, aes(sqft_living, price)) + geom_point(aes(color = rebuilt))
```

It looks like that rebuilt houses are a little more expensive than non-rebuilt houses. Note that this plot is logarithmic so being a little higher means a lot in real.

Let us confirm this by drawing the trend line.

```r
ggplot(rentals_for_rebuilt, aes(sqft_living, price)) + geom_smooth(aes(color = rebuilt))
```

As you can see the trendline for rebuilt houses is above the non rebuilt houses. So they are more expensive.
